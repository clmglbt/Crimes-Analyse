```python
# En quelle année y'a t'il eu le plus grand nombres de crimes ? 
crime.groupby('YEAR').count()['INCIDENT_NUMBER'].plot(kind='bar');
plt.title('Nombres de crimes');
```

```python
# Y a-t-il plus de crimes à des heures spécifiques ?
crime.groupby('HOUR').count()['INCIDENT_NUMBER'].plot(kind='bar');
```

```python
# Quels jours et à quelles heures les crimes sont-ils le plus souvent commis ?

week_and_hour = crime.groupby(['HOUR','DAY_OF_WEEK']).count()['INCIDENT_NUMBER'].unstack()

week_and_hour = week_and_hour[['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']]

sns.heatmap(week_and_hour, cmap=sns.cubehelix_palette(as_cmap=True));
```

```python
avg_crime = crime.groupby(['YEAR', 'MONTH']).count()['INCIDENT_NUMBER'].mean()
print("Le nombre moyen de crimes est de " + str(avg_crime))

year_and_month = crime.groupby(['MONTH', 'YEAR']).count()['INCIDENT_NUMBER'].unstack()

def style_negative(v, props=''):
    return props if v < avg_crime else None

def style_opacity(v):
    return 'opacity: 20%;' if (v < 0.3) and (v > -0.3) else None

s2 = year_and_month.style\
                .format(None, na_rep="")\
                .applymap(lambda v: 'color:red;' if v < avg_crime else None)\
                .applymap(style_opacity)
s2


def highlight_max(s, props=''):
    return np.where(s == np.nanmax(s.values), props, '')
s2.apply(highlight_max, props='color:white;background-color:green', axis=0)
```

```python
# Dans quels districts a-t-on commis le plus de crimes chaque année ?
district_and_year = crime.groupby(['DISTRICT', 'YEAR']).count()['INCIDENT_NUMBER'].unstack()

sns.heatmap(district_and_year, cmap=sns.cubehelix_palette(as_cmap=True));
```

```python
# Moyenne des crimes par district par année 

avg_crime_district = crime.groupby(['DISTRICT', 'YEAR']).count()['INCIDENT_NUMBER'].mean()
print("La moyenne des crimes par district par année est de : " + str(avg_crime_district))

def style_negative(v, props=''):
    return props if v < avg_crime_district else None
s3 = district_and_year.style.applymap(style_negative, props='color: orange;')\
              .applymap(lambda v: 'opacity: 20%;' if (v < 0.3) and (v > -0.3) else None)
s3
```
